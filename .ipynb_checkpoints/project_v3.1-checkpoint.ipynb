```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def create_array(num,df):
    array = []
    i = num
    for i in range(i,i+20):
        array.append(df['aX'].iloc[i])
    i = num
    for i in range(i,i+20):
        array.append(df['aY'].iloc[i])
    i = num
    for i in range(i,i+20):
        array.append(df['aZ'].iloc[i])
    i = num
    for i in range(i,i+20):
        array.append(df['gX'].iloc[i])
    i = num
    for i in range(i,i+20):
        array.append(df['gY'].iloc[i])
    i = num
    for i in range(i,i+20):
        array.append(df['gZ'].iloc[i])
    return array
```

```python
def fill_data(size,df):
    df_new = pd.read_csv('test.csv')
    i=0
    j=0
    while i <= size:
        df_new = pd.DataFrame(np.insert(df_new.values,j,values=create_array(i,df),axis=0))
        i = i + 20
        j = j+1
    return df_new
```

```python
def record_capture(capture):
    df_test = pd.read_csv(capture+'.csv')
    capt = fill_data(len(df_test)-1,df_test)
    capt = np.asarray(capt).astype('float32')
    
    return capt
```

```python
def final_res(result):
    for i in range(len(result)):
        if (np.rint(result[i][0]) == 1) & (np.rint(result[i][1]) == 1):
            print('forehand topspin')
        if (np.rint(result[i][0]) == 1) & (np.rint(result[i][1]) == 0):
            print('forehand slice')
        if (np.rint(result[i][0]) == 0) & (np.rint(result[i][1]) == 1):
            print('backhand topspin')
        if (np.rint(result[i][0]) == 0) & (np.rint(result[i][1]) == 0):
            print('backhand slice')
```

```python
def accuracy(x1,x2):
    error = 0
    for i in range(len(x1)):
        if (x1[i][0] != x2[i][0]) | (x1[i][1] != x2[i][1]) :
            error = error + 1
    acc = 100 - (error/len(x1))*100
    return round(acc,2)
```

```python
shots = pd.read_csv('shots.csv')
shots
```

```python
from sklearn.utils import shuffle
```

```python
shots = shuffle(shots)
```

```python
X = shots.drop(columns=['forehand','topspin','bad form'],axis=1).values
```

```python
y = shots[['forehand','topspin','bad form']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scalar.fit_transform(X_train)
```

```python
scalar.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model = Sequential()

model.add(Dense(units=120,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=60,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=20,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=3,activation='sigmoid'))

model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=1100,
          validation_data=(X_test, y_test),verbose=1,
          callbacks=[early_stop]
          )
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
predictions = model.predict(X_test)
```
